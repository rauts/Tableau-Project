The bar graph above shows that the female customers are more likely to leave the bank compared to male customers, `25 % vs 16 %`. The main question is **can we make a strong statement just by visual information?** Of course not, one needs to perfrom a statistical test to validate this statement. Since gender is a categorical variable, I use a 2x2 chi-square test for independance using stat package in Python.

__Prepare data for Chi-Square Test__

```python
# create cross tabble using two columns gender and exited from the dataframe.
Exited = data_ChurnMod['Exited'].replace({1: "Left", 0: "Stayed"}).astype('category')
Gender = data_ChurnMod['Gender'].astype('category')
# creare cross table 
cross_tab = pd.crosstab(Exited, Gender)
# using stat module 
cross_tab
```

__Hypothesis Testing__

The null hypothesis states that two groups (`Gender` and `Exited`) are independent, i.e., `Gender` is not associated with the `Exited` status (leaving the bank after 6 months).

Alternative hypothesis states that there is some kind of association between these two varialbes.

```python
# Load Statistical Module
from scipy.stats import chi2_contingency 
```

```python
chi2, p, dof, expected = chi2_contingency(cross_tab)
# p-value
print(f'p-value = {p}, degree-of-freedom = {dof}')
```

The p-value = 0.0001 which is very very less then 5% significance level indicates that there is an evidence that the null hypotheis shoould be rejected. In other words, data shows association between `Gender` and `Exited` status of customers. 
Interesting result, ist't it?
